```python
# Aufgabe 2
p  = (2,1,1)
v = (1,1)

def ExecutePerceptron(perceptron, values):
    result = 0
    for i in range(len(perceptron)):
        w = perceptron[i]
        if i == len(perceptron)-1:
            result += w
        else:
            result += w * values[i]
    return result

ExecutePerceptron(p,v)
```

### XOR
XOR kann nicht implementiert werden, da im Perceptron nur eine einzelne Trenngerade gibt, gäbe es 2 Trenngeraden könnte man es daarstellen.

```python
#Aufgabe 3
import random

#Datensatz
dataset = [(False, (random.uniform(-1,1),random.uniform(-1,1))) for _ in range(1000)]

#Gerade
g = (random.uniform(-1,1),random.uniform(-1,1),random.uniform(-1,1))

for i in range(len(dataset)):
    print(dataset[i])
    if ExecutePerceptron(g, dataset[i][1]) > 0:
        dataset[i] = (True, dataset[i][1])
dataset
```

```python
def TrainPerceptron(perceptron, dataset, maxIterations, learningRate):
    for generation in range(maxIterations):
        misclassified = []
        for d in dataset:
            res = True
            if ExecutePerceptron(perceptron, d[1]) < 0: res = False
            if res != d[0]:
                misclassified.append(d)
        if len(misclassified) == 0:
            return (perceptron, generation)
        
        r = random.choice(misclassified)
        pred = ExecutePerceptron(perceptron, r[1])
        if r[0]:
            error = 1 - pred
        else:
            error = -1 -pred
        
        percList = list(perceptron)
        for i in range(len(r[1])):
            percList[i] += learningRate * error * r[1][i]
        percList[-1] += learningRate * error
        perceptron = tuple(percList)
    return (perceptron, maxIterations)

print(g)
p = (0,0,0)
print(p)
print(TrainPerceptron(p,dataset,1000,1))
```

```python
#Training
max = -1
min = -1
sum = 0
for _ in range(1000):
    value = TrainPerceptron(p,dataset,1000,0.1)[1]
    if(value > max):
        max = value
    if(value < min or min ==-1):
        min = value
    sum += value
print(f"trained 1000 times, average Steps: {sum/1000}, minimum steps: {min}, maximum steps: {max}")
```
