```python
# Aufgabe 2
p  = (2,1,1)
v = (1,1)

def ExecutePerceptron(perceptron, values):
    result = 0
    for i in range(len(perceptron)):
        w = perceptron[i]
        if i == len(perceptron)-1:
            result += w
        else:
            result += w * values[i]
    return result

ExecutePerceptron(p,v)
```

XOR kann nicht implementiert werden, da im Perceptron nur eine einzelne Trenngerade gibt, gäbe es 2 Trenngeraden könnte man es daarstellen.

```python
#Aufgabe 3
import random

#Datensatz
dataset = [(False, (random.uniform(-1,1),random.uniform(-1,1))) for _ in range(10)]

#Gerade
g = (random.uniform(-1,1),random.uniform(-1,1),random.uniform(-1,1))

for d in dataset:
    print(d)
    if ExecutePerceptron(g, d[1]) > 0:
        d[0] = True
dataset
```

```python
def TrainPerceptron(perceptron, dataset, maxIterations, learningRate):
    for _ in range(maxIterations):
        misclassified = []
        for d in dataset:
            res = True
            if ExecutePerceptron(perceptron, d[1]) < 0: res = False
            if res != d[0]:
                misclassified.append(d)
        if len(misclassified) == 0:
            return perceptron
        r = random.choice(misclassified)
        print(r)
        pred = ExecutePerceptron(perceptron, r[1])
        error = r[0] - pred
        for i in range(len(r[1])):
            perceptron[i] += learningRate * error * r[1][i]
        perceptron[-1] += learningRate * error
    return perceptron

print(g)
p = (random.uniform(-1,1),random.uniform(-1,1),random.uniform(-1,1))
print(p)
print(TrainPerceptron(p,dataset,1,0.01))
```
